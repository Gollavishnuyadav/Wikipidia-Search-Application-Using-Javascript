```python
def print_pattern(n):
    prev = [0] * n
    curr = [0] * n

    # loop through each row
    for i in range(n):

        # print spaces before numbers
        for j in range(n-i-1):
            print(" ", end="")

        # loop through each number in the row
        for j in range(i+1):

            # calculate the value of the number
            if j == 0 or j == i:
                val = 1
            else:
                val = prev[j-1] + prev[j]

            # print the number
            print(val, end=" ")

            # store the value for the next row
            if i > 0:
                prev = curr.copy()
                curr[j] = val

        # move to the next row
        print()
print_pattern(4)
```

```python
def printPascal(n) :
    for line in range(0, n) :
        for i in range(0, line + 1) :
            print(binomialCoeff(line, i),
                " ", end = "")
        print()
 
def binomialCoeff(n, k) :
    res = 1
    if (k > n - k) :
        k = n - k
    for i in range(0 , k) :
        res = res * (n - i)
        res = res // (i + 1)
     
    return res
 
n = 5
printPascal(n)
```
