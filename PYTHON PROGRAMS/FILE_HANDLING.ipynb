```python
try:
    n=int(input())
    if n%2!=0:
        raise Exception("Not Even")
except Exception as e:
    print(e)
else:
    print("Even")
```

```python
class sample(Exception):
    def __init__(self,msg):
        self.msg=msg
    def __str__(self):
        return self.msg
try:
    n=int(input())
    if n%2!=0:
        raise sample('Not Even')
except Exception as e:
    print(e)
else:
    print("Even")
```

```python
class Prime(Exception):
    def __init__(self,msg):
        self.msg=msg
    def __str__(self):
        return self.msg
try:
    n=int(input())
    for i in range(2,n//2+1):
        if n%2==0:
            raise Prime('Not Prime')
        
except Prime as e:
    print(e)
else:
    
    print("Prime")
finally:
    print("Done")
```

```python
#assert is a keyword .it check the condtion,if the given condtion is false it through and asserion error
try:
    n=int(input())
    assert n%2==0,'Not Even'
except AssertionError as e:
    print(e)
```

#GENERATOR

```python
#GENERATOR IS A NORMAL FUNCTION WITH YIELD KEYWORD
#OUTPUT OF GENERATOR FUNCTION IS GENARATOR OBJECT
#GENERATOR OBJ USED EITHER ON NEXT OR FOR LOOP
def one():
    yield 1
    yield 2
    yield 3
    yield 4
x=one()
for i in x:
    print(i)
```

```python
def fact(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
        yield fact
x=fact(5)
print(next(x))
for i in x:
    print(i)
```
