```python
def find_second_highest(lst):
    if len(lst) < 2:
        return "List should have at least two elements"
    
    max_val = lst[0]
    second_max = float('-inf')

    for num in lst:
        if num > max_val:
            second_max = max_val
            max_val = num
        elif num > second_max and num != max_val:
            second_max = num

    if second_max == float('-inf'):
        return "There is no second highest value"
    else:
        return second_max

# Example usage
my_list = [5, 10, 3, 8, 2]
second_highest = find_second_highest(my_list)
print("Second highest value:", second_highest)
```

```python
#start()
#start help the execution of threads
#join() join help to termination of thread
#is_alive() This is is_alive() method check whether the thread is still executing or not
#name() help to print name of thread

#threading.active_count()--return number of thread objects that are active
#threading.enumerate()-->Return list of all thread objects that are currently active

import time
import threading
def fun1(n):
    for i in range(n):

        print('thread One')
        time.sleep(2)

def fun2(n):
     for i in range(n):
        print('thread TWO')
        time.sleep(2)
t1=time.time()
th=threading.Thread(target=fun1,args=(5,),name='Golla')
th2=threading.Thread(target=fun2,args=(5,),name="Vishnu")
th.start()
th2.start()
th.join()
th2.join()
threading.activeCount()
threading.enumerate()
threading.main_thread()

t2=time.time()
t3=t2-t1
print(t3)
```

```python
import unittest

def add_numbers(x, y):
    return x + y

class TestAddNumbers(unittest.TestCase):
    def test_add_positive_numbers(self):
        result = add_numbers(5, 10)
        self.assertEqual(result, 15)

    def test_add_negative_numbers(self):
        result = add_numbers(-5, -10)
        self.assertEqual(result, -15)

    def test_add_zero_to_number(self):
        result = add_numbers(10, 0)
        self.assertEqual(result, 10)

if __name__ == '__main__':
    unittest.main()
```
