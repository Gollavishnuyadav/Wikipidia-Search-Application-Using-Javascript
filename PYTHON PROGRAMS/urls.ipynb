```python
from weatherapp.models import City  # Importing the City model from weatherapp.models
from django.shortcuts import get_object_or_404, redirect, render  # Importing necessary shortcuts
from decouple import config  # Importing the decouple library to manage environment variables
import requests  # Importing the requests library for making HTTP requests
from pprint import pprint  # Importing pprint for pretty printing data (though it's not used in this code)
from django.contrib import messages  # Importing messages to display user notifications


# Import necessary modules and classes
from weatherapp.models import City
from django.shortcuts import get_object_or_404, redirect, render
from decouple import config
import requests
from pprint import pprint
from django.contrib import messages

# Define the home view function
def home(request):
    # Set the OpenWeatherMap API key
    API_key = '30d4741c779ba94c470ca1f63045390a'
    
    # Get the user-entered city name from the query parameter 'name'
    u_city = request.GET.get('name')
    if u_city:
        # Construct the API URL for the specified city
        url = f"https://api.openweathermap.org/data/2.5/weather?q={u_city}&appid={API_key}&units=metric"
        
        # Send a GET request to the API
        response = requests.get(url)
        
        # Check if the response is successful
        if response.ok:
            # Parse the JSON content of the response
            content = response.json()
            
            # Extract the city name from the API response
            r_city = content['name']
            
            # Check if the city already exists in the database
            if City.objects.filter(name=r_city):
                # Display a warning message if the city exists
                messages.warning(request, 'City already exists!')
            else:
                # Create a new City object in the database
                City.objects.create(name=r_city)
                
                # Display a success message for city creation
                messages.success(request, 'City created!')
        else:
            # Display an error message if the city couldn't be found in the API
            messages.error(request, 'There is no city!')
        
        # Redirect back to the home view after processing
        return redirect('home')
    
    # Prepare a list to store weather data for cities
    city_data = []
    
    # Get all City objects from the database
    cities = City.objects.all()
    
    # Loop through each city to fetch weather data
    for city in cities:
        # Construct the API URL for the current city
        url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_key}&units=metric"
        
        # Send a GET request to the API
        response = requests.get(url)
        
        # Parse the JSON content of the response
        content = response.json()
        
        # Extract relevant weather data and create a dictionary
        data = {
            'city_api': content['name'],  # Store the API-provided city name using 'city_api'
            'city_db': city,              # Store the City object from the database using 'city_db'
            'temp': content['main']['temp'],
            'icon': content['weather'][0]['icon'],
            'desc': content['weather'][0]['description'],
        }
        
        # Add the data to the city_data list
        city_data.append(data)
        
        # Print the city_data for debugging purposes
        print(city_data)
        
    # Prepare the context dictionary for rendering the template
    context = {
        'city_data': city_data,
    }
    
    # Render the 'weatherapp/home.html' template with the context
    return render(request, 'weatherapp/home.html', context)

# Define the delete city view function
def delete_city(request, id):
    # Get the City object based on the provided id, or raise a 404 error if not found
    city = get_object_or_404(City, id=id)
    
    # Delete the retrieved City object from the database
    city.delete()
    
    # Display a success message for city deletion
    messages.success(request, 'City deleted!')
    
    # Redirect back to the home view after deleting the city
    return redirect('home')
```
