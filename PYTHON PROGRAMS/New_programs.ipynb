```python
def is_perfect_number(num):
    divisors = []
    for i in range(1, num):
        if num % i == 0:
            divisors.append(i)

    if sum(divisors) == num:
        return True
    else:
        return False

perfect_numbers = []

for num in range(10, 101):
    if is_perfect_number(num):
        perfect_numbers.append(num)

# Print the perfect numbers
for perfect_num in perfect_numbers:
    print(perfect_num)
```

```python
# Python3 code to demonstrate working of
# Exceptional Split in String
# Using loop + split()

# initializing string
test_str = "gfg, is, (best, for), geeks"

# printing original string
print("The original string is : " + test_str)

# Exceptional Split in String
# Using loop + split()
temp = ''
res = []
check = 0
for ele in test_str:
	if ele == '(':
		
		check += 1
		print(check)
	elif ele == ')':
		check -= 1
		print(check)
	if ele == ', ' and check == 0:
		if temp.strip():
			res.append(temp)
		temp = ''
	else:
		temp += ele
if temp.strip():
	res.append(temp)

# printing result
print("The string after exceptional split : " + str(res))
```

```python
class BookMyShow:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
            cls._instance.movies = {}  # Store movies and their details
        return cls._instance

    def add_movie(self, movie_name, available_seats):
        self.movies[movie_name] = {'available_seats': available_seats}

    def book_ticket(self, movie_name, num_tickets):
        if movie_name in self.movies and self.movies[movie_name]['available_seats'] >= num_tickets:
            self.movies[movie_name]['available_seats'] -= num_tickets
            print(f"{num_tickets} tickets booked for {movie_name}")
        else:
            print("Sorry, no available seats for the selected movie.")

# Usage
if __name__ == "__main__":
    book_my_show = BookMyShow()

    book_my_show.add_movie("Avengers: Endgame", 50)
    book_my_show.add_movie("Spider-Man: Far From Home", 30)

    book_my_show.book_ticket("Avengers: Endgame", 3)
    book_my_show.book_ticket("Spider-Man: Far From Home", 2)
```
