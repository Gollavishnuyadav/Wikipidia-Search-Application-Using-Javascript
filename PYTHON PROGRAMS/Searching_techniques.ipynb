```python
class Bank:
    """THIS BANK BELONGS TO ICICI"""
    Bname='ICICI'
    Roi=.01
    def __init__(self,name,bal,pin):
        self.name=name
        self.bal=bal
        self.pin=pin
    def CheckBalance(self):
        if self.pin==self.getpin():
            print(f'{self.name}{self.bal}')
        else:
            print('Incorrect password')
    def Withdraw(self,amt):
        if self.pin==self.getpin():
            if self.bal>=amt:
                self.bal-=amt
                print('debited successfully')
            else:
                print('Insufficient Balance')
        else:
            print('Incorrect Pin')
    def Deposited(self,amt):
        if self.pin==self.getpin():
            self.bal+=amt
            print(f'{self.bal}')
        else:
            print('Incorrect pin')
    @classmethod
    def changeROI(cls,n):
        cls.Roi=n
    @staticmethod
    def getpin():
        return int(input())
obj=Bank('Golla',20000,1234)
obj.__doc__
obj.Deposited(2000)
obj.CheckBalance()
```

```python
def prime(n):
    if n<=1:
        return False
    for i in range(2,n//2+1):
        if n%2==0:
            return False
    else:
        return True
l=[1,2,3,4,5,6,7,8,9]
xx=list(filter(prime,l))
xx
```

```python
def even(n):
    if n%2==0:
        return n
l=[1,2,3,4,5,6,7,8]
xx=list(filter(even,l))
xx
```
