```python
class BankAccount:
    def __init__(self, account_number):
        self.account_number = str(account_number)
        self.balance = 0

    def get_balance(self):
        return self.balance

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            raise ValueError("Insufficient Funds")

    def deposit(self, amount):
        self.balance += amount


def transfer_amount(acc_1, acc_2, amount):  
    try:
        acc_1.withdraw(amount)
        acc_2.deposit(amount)
        return True
    except ValueError as e:
        print(str(e))  
        print(type(e))  
        print(e.args)  
        return False

user_1 = BankAccount("001")
user_2 = BankAccount("002")
user_1.deposit(25)
user_2.deposit(100)

print("User 1 Balance: {}/-".format(user_1.get_balance()))
print("User 2 Balance: {}/-".format(user_2.get_balance()))
print(transfer_amount(user_1, user_2, 50))
print("Transferring 50/- from User 1 to User 2")
print("User 1 Balance: {}/-".format(user_1.get_balance()))
print("User 2 Balance: {}/-".format(user_2.get_balance()))
```

```python
class seq():
    def __init__(self,start,end):
        self.start=start
        self.end=end
    def __iter__(self):
        return self
    def __next__(self):
        n=self.start
        if self.start>=self.end:
            raise StopIteration
        self.start+=2
        return n
xx=seq(1,10)
ob=iter(xx)
for i in ob:
    print(i)
```

```python
def fact(n):
    fact=1
    for  i in range(1,n+1):
        fact=fact*i
        yield fact
x=fact(5)
for i in x:
    print(i)
```

```python
def fib(n):
    f,s=0,1
    for i in range(n):
        t=f+s
        yield f
        f,s=s,t
x=fib(5)
for i in x:
    print(i)
```

```python
# Standard dict
my_dict = {'apple': 1, 'banana': 2, 'orange': 3}

# defaultdict with default value of 0
from collections import defaultdict
my_default_dict = defaultdict(int)
my_default_dict['apple'] += 1
print(my_default_dict['apple'])  # Output: 2
print(my_default_dict['pear'])   # Output: 0 (default value)

# OrderedDict
from collections import OrderedDict
my_ordered_dict = OrderedDict()
my_ordered_dict['apple'] = 1
my_ordered_dict['banana'] = 2
my_ordered_dict['orange'] = 3
for fruit, count in my_ordered_dict.items():
    print(fruit, count)
# Output:
# apple 1
# banana 2
# orange 3
```

```python
def is_prime(n, i=2):
    if n < 2:
        return False
    elif n == 2 or n == 3:
        return True
    elif n % i == 0:
        return False
    elif i * i > n:
        return True
    else:
        return is_prime(n, i + 1)
is_prime(9)
```
