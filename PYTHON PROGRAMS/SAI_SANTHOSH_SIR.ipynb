```python
import csv
import json

# The JSON data
json_data = [
    {
        "id": 1,
        "data": {
            "seriesy": 10,
            "color": "#FF0000",
            "category": "chart1",
            "seriesx": 1,
            "subtitle": "Subtitle 1",
            "title": "Title 1"
        }
    },
    {
        "id": 2,
        "data": {
            "seriesy": 20,
            "color": "#00FF00",
            "category": "chart1",
            "seriesx": 2,
            "subtitle": "Subtitle 2",
            "title": "Title 2"
        }
    },
    {
        "id": 3,
        "data": {
            "seriesy": 30,
            "color": "#0000FF",
            "category": "chart1",
            "seriesx": 3,
            "subtitle": "Subtitle 3",
            "title": "Title 3"
        }
    }
]

# The CSV file to write to
csv_file = 'output.csv'

# Open the CSV file for writing
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)

    # Write the header
    writer.writerow(["id", "seriesy", "color", "category", "seriesx", "subtitle", "title"])

    # Write the data
    for item in json_data:
        id = item["id"]
        data = item["data"]
        writer.writerow([
            id,
            data["seriesy"],
            data["color"],
            data["category"],
            data["seriesx"],
            data["subtitle"],
            data["title"]
        ])

print(f'Data has been written to {csv_file}')
```
